```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_style("darkgrid")

# Resource File (CSV)
df = pd.read_csv('../Project1_exploration/output_data/Top_497.csv')
df.head()
```

```python
#Reorganising columns so Cuisine is at the end of the dataframe - dropping unnamed.
cuisines = df[['Restaurant_id',"Name", "Locality", "Address", "City", "Latitude", "Longitude", "Price Range", 
                      "Average Cost for two", "User Rating", "Rating Text", "Votes", "Cuisines"]]
cuisines
```

```python
#Separating Cuisines data by comma and placing the data into new columns
cuisines = pd.concat([cuisines, cuisines['Cuisines'].str.split(', ', expand=True)], axis=1)
cuisines.columns = ['Restaurant_id',"Name", "Locality", "Address", "City", "Latitude", "Longitude", "Price Range", 
                      "Average Cost for two", "User Rating", "Rating Text", "Votes", "Cuisines", "Cuisine_1","Cuisine_2","Cuisine_3","Cuisine_4","Cuisine_5","Cuisine_6","Cuisine_7","Cuisine_8"]
cuisines
```

```python
#Placing Cuisine_1 the main cuisine in a Dataframe to allow grouping
cuisines1 = cuisines["Cuisine_1"].value_counts()
cuisines1
```

```python
# Combining similar cuisines together
cuisines = cuisines.replace({"Sushi": "Japanese", "Japanese BBQ": "Japanese", "Ramen" : "Japanese",
                             "Pizza": "Mediterranean", "European/Mediterranean": "Mediterranean", "Italian": "Mediterranean",
                             "Greek" : "Mediterranean", "Spanish/Tapas": "Mediterranean", "Tapas" : "Mediterranean",
                             "Spanish": "Mediterranean","Portuguese":"Mediterranean",
                             "European": "Other European", "Modern European":"Other European", "French": "Other European",
                             "Crepes": "Other European", "British": "Other European","German": "Other European",
                             "Lebanese": "Middle Eastern","Turkish":"Middle Eastern",
                             "Bar Food": "Bar/Pub", "Pub Food": "Bar/Pub",
                             "Cafe Food": "Cafe/Bakery", "Bakery": "Cafe/Bakery", "Sandwich": "Cafe/Bakery",
                             "Modern Australian": "Australian", 
                             "Burger" : "American", "American/Burger": "American", 
                             "Asian": "Other Asian", "Asian Fusion": "Other Asian", "Asian Other": "Other Asian",
                             "Korean BBQ":"Korean", "Tibetan": "Other Asian","Laotian": "Other Asian", "Cantonese": "Chinese",
                             "Fast Food":"American", 
                             "Argentine": "Latin American","Mexican": "Latin American",
                             "Vegan": "Vegan/Vegetarian", "Vegetarian": "Vegan/Vegetarian",
                             "Coffee and Tea": "Other", "Desserts":"Other", "Seafood": "Other", "BBQ": "Other", 
                             "Steak": "Other", "Healthy Food": "Other", "Fish and Chips": "Other", "Ice Cream": "Other", 
                             "Contemporary": "Other"
         })

cuisines2 = cuisines["Cuisine_1"].value_counts()
cuisines2
```

```python
#Keep the top ten cuisines we are interested in:
final_df = cuisines.loc[(df["Cuisine_1"] == "Japanese") |
                                  (df["Cuisine_1"] == "Asian") |
                                  (df["Cuisine_1"] == "Italian")|
                                   (df["Cuisine_1"] == "Australian") |
                                   (df["Cuisine_1"] == "European") |
                  (df["Cuisine_1"] == "American/Burger") |
                  (df["Cuisine_1"] == "Indian") |
                  (df["Cuisine_1"] == "Vietnamese") |
                  (df["Cuisine_1"] == "Spanish/Tapas") |
                  (df["Cuisine_1"] == "Middle Eastern") 
                                  ]
final_df["Cuisine_1"]
```

```python
city_group = final_df.groupby("City")
cuisine_city = pd.DataFrame(city_group["Cuisine_1"].value_counts())
cuisine_city.columns = ["Count"]
cuisine_city
```

```python
#Breakdown of restaurants per user ratings across cities
cuisine=final_df.groupby(['City'])['Cuisine_1'].value_counts().unstack('Cuisine_1').fillna(0)
cuisine.plot(kind='bar', stacked=True, legend=False, rot=0, figsize=(10,5)).legend(bbox_to_anchor=(1, 1))
# Set textual properties
plt.title("Restaurants per cuisines per city")
plt.ylabel("Number of restaurants")
plt.xlim(-0.5,len(cuisine)-0.5)
plt.ylim(0,72)
#Save image
plt.savefig('images/Restaurants per cuisines per city.png')
# Show plot
plt.show()
plt.tight_layout()
```

```python
cuisine_group = final_df.groupby("Cuisine_1")
cities_per_cuisines = pd.DataFrame(cuisine_group["City"].value_counts())
cities_per_cuisines.columns = ["Count"]
cities_per_cuisines
```

```python
#Breakdown of restaurants per user ratings across cities
cuisine2=final_df.groupby(['Cuisine_1'])['City'].value_counts().sort_values(ascending=False).unstack('City').fillna(0)
cuisine2.plot(kind='barh',stacked=True, legend=False, figsize=(15,5)).legend(bbox_to_anchor=(1, 1))
# Set textual properties
plt.title("Restaurants per city per cuisines")
plt.ylabel("Number of restaurants")
plt.ylim(-0.5,len(cuisine2)-0.5)
plt.xlim(0,85)
#Save image
plt.savefig('images/Restaurants per city per cuisines.png')
# Show plot
plt.show()
plt.tight_layout()
```

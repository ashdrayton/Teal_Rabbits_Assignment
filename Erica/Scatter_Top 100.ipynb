```python
#Import required modules and styles
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("darkgrid")
```

```python
#Read in required data, drop first column, sort by city, filter columns and reset the index
df = pd.read_csv('../Project1_exploration/output_data/Top_100.csv')
df.drop('Unnamed: 0', axis=1)
df.sort_values(by='City', inplace=True)
cols = ['Name', 'City', 'Average Cost for two', 'User Rating']
restaurants_100 = df.loc[:, cols]
restaurants_100 = restaurants_100.reset_index(drop=True)
restaurants_100.to_csv('restaurants100.csv')
```

```python
df = pd.read_csv('restaurants100.csv')
```

```python
#Define variables
x1 = df.iloc[0:7, 3]
y1 = df.iloc[0:7, 4]
x2 = df.iloc[8:17, 3]
y2 = df.iloc[8:17, 4]
x3 = df.iloc[18:61, 3]
y3 = df.iloc[18:61, 4]
x4 = df.iloc[62:73, 3]
y4 = df.iloc[62:73, 4]
x5 = df.iloc[74:99, 3]
y5 = df.iloc[74:99, 4]
```

```python
#Specify plot to be a scatter plot and add variables
plt.scatter(x1,y1, label='Adelaide')
plt.scatter(x2,y2, label='Brisbane')
plt.scatter(x3,y3, label='Melbourne')
plt.scatter(x4,y4, label='Perth')
plt.scatter(x5,y5, label='Sydney')

#Set plot x and y axis limits
plt.axis([0, 700, 4.6, 5.0])

#Adjust size of graph
fig = plt.gcf()
fig.set_size_inches(15, 12)

#Set font sizes
small = 12
medium = 16
large = 24

plt.rc('font', size=small)     
plt.rc('axes', titlesize=small)     
plt.rc('axes', labelsize=small) 
plt.rc('xtick', labelsize=small)   
plt.rc('ytick', labelsize=small)   
plt.rc('legend', fontsize=small)  
plt.rc('figure', titlesize=large)

#Add legend and set graph and legend position
#plt.legend(loc='best')
plt.subplots_adjust(left=0.15, bottom=0.3)
plt.legend(bbox_to_anchor=(-0.05, 0.65), loc='upper right')

#Add title, axis labels and gridlines
plt.title('Comparison of User Rating vs Average Cost for Two People')
plt.xlabel('Average Cost for Two People')
plt.ylabel('User Rating')
plt.grid(which='major', linestyle='-', color='black')

#Save plot
plt.savefig("scatter_100.png")
```
